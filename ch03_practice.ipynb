```python
# 4-23 利用辗转相除法求两个整数之间的最大公约数和最小公约数
# 分析：利用大数除以小数，若余数不为0，则将小数送给大数，余数送给小数；
#       余数为0时，小数变量中的数即为最大公约数。
#       最小公倍数等于两数乘积除以最大公约数。
a,b= eval(input("请输入两个整数,逗号分隔开:"))
max_num = max(a,b)
min_num = min(a,b)
while max_num%min_num !=0:
    min_num_copy = min_num
    min_num = max_num%min_num
    max_num = min_num_copy

A = min_num
print("最大公约数是:%d"%(A))
print("最小公倍数是:%d"%(a*b/A))
```

```python
# 4-24 利用莱布尼茨公式求出圆周率
epsilon=1e-7
n = 1
item = 1
pi_4=0
while abs(item)>=epsilon:
    pi_4 += item
    n=n+1
    item = ((-1)**(n-1))*(1/(2*n-1))

print("圆周率为%f"%(pi_4*4))
```

```python
# 4-25 将一个正整数分解质因数
#解题思路：
#设待分解的数为x，分解质因数过程可按以下步骤进行：
# （1）首先找到x的一个最小质因数k(2~x)；
# （2）若这个质因数k等于x，则说明分解结束，跳出循环，打印最后一个质因数x；
# （3）若这个质因数k不等于x，但因x能被k整除，所以k是x的一个质因数，应打印出k的值
#      并用x整除k的商作为新的正整数x，重复执行第（1）步
n = int(input("请输入一个正整数:"))
if n == 1:
    print("1没有质因数")
else:
    k=2
    while k<=n: #一个一个数字尝试
        if n%k ==0: #如果可以整除
            if n==k: #如果他是这个数字本身 那就停止了
                print(k)
                break
            else:
                print("%dx"%(k),end="") #如果没有那就以这个格式进行输出
                n=n//k  # 除掉他 继续开始算
        else:           # 如果没有 那就跳到下一个数字
            k+=1
```

```python
# 4-26 输出菱形图形(见PPT)
# 分析:利用两个嵌套循环。第一个双重循环控制前6行输出，
#      每行先输出递减个数的空格，然后再输出*；
#      第二个双重循环控制后5行输出，
#      每行先输出递增个数的空格，然后再输出*。每行结束要换行
for i in range(6):
    for _ in range(6-i):
        print(" ",end="")
    for _ in range(2*i+1):
        print("*",end="")
    print("")

for j in range(5):
    for _ in range(j+2):
        print(" ",end="")
    for _ in range(-2*j+9):
        print("*",end="")
    print("")
    
```
