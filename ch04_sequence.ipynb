```python
# 4.1 列表
# 创建与访问

# 1.创建方法
# 基本的创建法就不说啦
# 使用list()函数把range对象、字符串、元组或其他可迭代对象转换成列表
list(range(1,11))
list("Python")
```

```python
list1 = list(range(1,11))
# 2.访问
# 首先是索引访问,其实就是数字访问
list1[1]
print(list1[-2]) #倒数第二个元素
# 如果索引超出了范围 就会报错

# 其次是切片访问[start:end:step]
# 从start开始，以step为步长，顺序取得列表中的元素，直到end-1位置
list1[1:10:2]

# 注意!!!
# 使用切片方式截取列表，如果下标出界，不会抛出异常，
# 而是在列表尾部截断或者返回一个空列表，代码具有更强的健壮性。
```

```python
# scores 表示语文、英语、数学、物理、化学和生物的成绩
scores = [75, 80, 85, 70, 65, 90]
scores[0]        # 语文成绩
scores[-1]       # 最后一门课成绩
scores[0:3]      # 语文英语数学成绩
scores[2:5]      # 数学物理化学成绩
scores[3:]       # 物理化学生物成绩。end为列表长度时可以省略
scores[0:6:2]    # 语文数学化学成绩
scores[10]       # 报错:索引出界
scores[0:10]     # 所有课程成绩
```

```python
# 4-1 计算一组成绩的最高分 最低分 和平均分
scores = [75, 80, 85, 70, 65, 90]
print('最高分：', max(scores))
print('最低分：', min(scores))
s = sum(scores)  # 总分
n = len(scores)

#要注意的是 python本体没有mean函数 别直接用
print('平均分：', round(s/n, 1))
```

```python
'''列表对象的方法'''
# 对象具有属性和方法 调用格式为:

# 对象.属性名
# 对象.方法名(参数)
# 3. 列表元素添加（append vs extend）
x1 = [1,2,3]
x1.append(5)        # 追加单个元素，结果：[1,2,3,5]
x1.append([7,8])    # 追加列表（作为单个元素），结果：[1,2,3,5,[7,8]]

x2 = [1,2,3]
x2.extend([7,8])    # 扩展列表（合并元素），结果：[1,2,3,7,8]

x3 = [1,5,7,2]
x3.reverse()        # 逆序排列，结果：[2,7,5,1]
x3.sort(reverse=True)  # 降序排序，结果：[7,5,2,1]
x3[3] = 6           # 修改索引3的元素，结果：[7,5,2,6]
x3.sort()           # 升序排序，结果：[2,5,6,7]

# 定义列表x4，包含重复元素和不同字符
x4 = ['a', 'b', 'a', 'b', 'c']

# index('a')：查找元素'a'在列表中首次出现的索引（从0开始计数）
# 结果：0（第一个元素就是'a'）
x4.index('a')

# pop()：默认删除列表最后一个元素，无参数时操作尾部
# 原列表['a','b','a','b','c'] → 删除最后一个'c' → 结果：['a', 'b', 'a', 'b']
x4.pop()
x4  # 此时列表状态：['a', 'b', 'a', 'b']

# pop(0)：删除指定索引（0）对应的元素，即列表第一个元素
# 上一步结果['a','b','a','b'] → 删除索引0的'a' → 结果：['b', 'a', 'b']
x4.pop(0)
x4  # 此时列表状态：['b', 'a', 'b']

# remove('b')：删除列表中首次出现的指定元素'b'（不是按索引，是按元素值）
# 上一步结果['b','a','b'] → 删除第一个'b' → 结果：['a', 'b']
x4.remove('b')
x4  # 此时列表状态：['a', 'b']

# clear()：清空列表中所有元素，列表变为空列表
# 上一步结果['a','b'] → 清空后 → 结果：[]
x4.clear()
x4  # 最终列表状态：[]
```
