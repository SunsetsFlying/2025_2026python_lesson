```python
# 标识符；
# 标识符指的就是在程序中你自己设置的变量、函数、类的名字
# 标识符只能由字母、数字、下划线组成，且必须由字母或下划线开头，不然会报错，以及不能和关键字同名
```

```python
# 常量：
# 整型常量：
# 十进制(%d)：123
# 八进制(%o)：以0o或者0O开头去表示: 0o123 换成十进制就是83
# 十六进制(%x)：以0x或者0X开头去表示: 0x123 就是十进制的291

# 浮点型常量(实型常量)：
# 常规形式:1.23 这是人看的懂的形式
# 机器一般用指数形式表示:3.0e8 表示 3x10的8次方

#字符型常量(字符串)
# 表示方法没什么好说的
# 主要了解比较特殊的字符型常量 转义字符 \
# 常用的比如 \n 表示换行 \\ 用以输出真正的反斜杠号"\" 还有类似的方法输入单引号和双引号
# 自己去记一下 比如 \101 表示把101转成八进制字符串 \0x101表示转成十六进制字符串

#布尔型常量
# 没什么好说的

#复数型常量
# 没什么好说的
```

```python
# 变量：
# 变量结构
# python中一切变量都称之为对象，变量的存储采用引用语义的方式
# 例子：我输入 a=1 python就会 在对象里用内存存入1  用指针把你命名的变量和对象连接起来，这就叫引用
    # 假如你又输入b=1 那么b就会被引用到与a同样的位置 因为他们的值是一样的

# 特殊变量：
# _xxx
# __xxx
# __xxx__
# 三种类型的变量，最好配以例子去体现三者之间的区别
```

```python
# 运算符：
#     特别多，分为：
#         算术运算符
#         关系运算符
#         逻辑运算符
#         赋值运算符
#         位运算符
#         成员运算符
#         身份运算符
```

```python
# 基本的输入和输出方法：
# print()函数
s = 'Python'
score = 87.5
print("你的%s是:%f" %(s,score))
# format()函数
# 通过{}代替以前的%
# 在大括号内的数字编号后，还可以添加更为详细的格式定义
# 数字编号与格式定义之间利用:去分开
print("你的{0}分数是:{1:.4f}".format(s,score))
# 格式化的字符串常量
#  f""这样子去写
# 同样可以在大括号内:分割开去设置详细的格式
print(f"你的{s}分数是:{score}")
```

```python
# input函数
# 例子
# 利用input函数从键盘里输入两个整数 计算两个数字之和
a = eval(input("请输入整数a:"))
b = eval(input("请输入整数b:"))
print(f"{a+b:.3f}")
```
