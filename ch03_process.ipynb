```python
# 例4-3 设三角形三条边边长分别为abc 三角形面积计算公式为?
# 利用 海伦公式
import math
a,b,c = eval(input("请输入三角形的三条边长abc,以逗号间隔:"))
s = (a+b+c)/2.0
S = math.sqrt(s*(s-a)*(s-b)*(s-c))
print(f"面积为{S:.2f}")
```

```python
#例4-4 将键盘输入的三位整数逆置输出
# 利用整除和求余计算分别将整数的个位 十位和百位取出 然后从高位到地位输出
x = int(input("请输入一个三位整数:"))
a = x//100
b = x//10%10
c = x%10
print("%d%d%d\n"%(c,b,a))   #逆置输出
```
