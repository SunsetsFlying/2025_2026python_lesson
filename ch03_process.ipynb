```python
# 例4-3 设三角形三条边边长分别为abc 三角形面积计算公式为?
# 利用 海伦公式
import math
a,b,c = eval(input("请输入三角形的三条边长abc,以逗号间隔:"))
s = (a+b+c)/2.0
S = math.sqrt(s*(s-a)*(s-b)*(s-c))
print(f"面积为{S:.2f}")
```

```python
'''单分支示例'''
x = int(input("请输入一个数字:"))
if x<0:
    x = -x
print("|x|=%d\n"%(x))
```

```python
'''双分支示例'''
# 例4-6 判断是否是闰年
# 判断方法:年份能否被4整除而不能被100整除 或者年份能被400整除
year = eval(input('请输入年份:'))
if (year%4==0 and year%100!=0) or year%400==0:
    print("年份%d是闰年\n"%(year))
else:
    print("年份%d不是闰年\n"%(year))
```

```python
'''多分支示例'''
#例4-7 判断一元二次方程的公式
import math
a,b,c=eval(input("请输入方程的各项系数,用逗号分隔"))
delta = b**2-4*a*c
if delta==0:
    print("两个相等实根为:%f\n"%(-b/(2*a)))
elif delta>0:
    x1 = (-b+math.sqrt(delta))/2*a
    x2 = (-b-math.sqrt(delta))/2*a
    print("x1:%f,x2=%f"%(x1,x2))
else:
    print("无实根")
```

```python
'''嵌套分支'''
# 例4-12 输入两个整数 比较二者之间的大小关系 利用嵌套分支
a,b = eval(input("请输入两个整数:")) 
if a!=b:
    if a>b:
        print("%d>%d"%(a,b))
    else:
        print("%d<%d"%(a,b))
else:
    print("%d和%d相等"%(a,b))
```

```python
'''for语句的例子'''
# 4-15 for循环求s=1+2+...+n
n = int(input('请输入正整数n=:'))
s=0
for i in range(1,n+1):
    s = s+i

print("s=%d\n"%(s))

```

```python
# 4-16 判断一个数字是否为素数
# 判断方法:一般只需要测试n是否能被2，3，…，n/2(或√n)整除，只要能被其中一个数整除，则n就不是素数
n = int(input("请输入正整数n="))
flag = 1
for i in range(2,n//2+1):
    if n%i==0:
        flag=0
        break

if flag==1:
    print("%d是素数"%(n))
else:
    print("%d不是素数"%(n))
```

```python
# 4-17 输出100-999里所有水仙花数(每个位的数字的立方和相加为该数字本身)
for i in range(100,1000):
    a=i//100
    b=i//10%10
    c=i%10
    if a**3+b**3+c**3==i:
        print(i,end=" ")
```

```python
'''continue语句的例子'''
for i in range(1,10):
    if i%5==0:
        continue
    else:
        print(i,end=" ")
```

```python
'''嵌套循环例子'''
# 4-19 输出九九乘法表
for i in range(1,10):
    for j in range(1,i+1):
        print("%dx%d=%d"%(i,j,i*j),end=" ")
    print("\n")
```

```python
# 4-20 输出100以内所有的素数 每行输出10个
n=0
str=""
print("100内的素数为:")
for i in range(2,101):
    flag=1
    for j in range(2,i//2+1):
        if i%j==0:
            flag=0
            break
    if flag==1:
        n+=1
        str=str+"%-2d"%(i)
        if n%10==0:
            print(str)
            str=""
print(str)
```

```python
'''一些综合案例'''
#冒泡排序:
#外循环循环n-1次
#内循环只在无序层遍历 只做一个操作:如果左边的数字比右边的大 那就交换位置 否则不变
list=[21,65,36,98,542,11,124,623,96,547,52,222]
new_list = list
for i in range(len(list)-1):
    for j in range(len(list)-1-i):
        if list[j+1]<list[j]:
            new_list[j+1],new_list[j] = list[j],list[j+1]
#降序排序只需要把小于号改成大于号就好啦~

```

```python
#选择排序:这个的逻辑很简单 每次抽出最小值，然后放到一个新的列表里
list=[21,65,36,98,542,11,124,623,96,547,52,222]
new_list=[]
for i in range(len(list)):  #循环n次
    minvalue = min(list)
    new_list.append(minvalue)
    list.remove(minvalue)
new_list
```

```python
epsilon = 10e-5
s = 0  # 累加和初始化为0
n = 1  # 项的索引从1开始

# 单个while循环：当当前项的绝对值 >= epsilon时继续累加
while abs(((-1)**(n-1)) * (1/(2*n - 1))) >= epsilon:
    term = ((-1)**(n-1)) * (1/(2*n - 1))  # 计算当前项
    s += term  # 累加当前项到和中
    n += 1  # 索引+1，准备下一项

# 最后需要把退出循环前的最后一项（已满足小于epsilon）也加上
s += ((-1)**(n-1)) * (1/(2*n - 1))

print("圆周率的值为%f\n" % (4 * s))
```
