```python
#第五章 --函数--

# 5.1 函数的定义域调用**
# 5.1 函数的定义域调用**
'''
1.函数的定义
def  函数名(参数1 参数2 参数3):
    
    [函数体]

1.括号里面的参数为形式参数，实际调用的时候运用的参数叫做实际参数
2.可以使用return语句返回函数代码的执行结果 返回值可以一个或者多个 若没有则默认返回None
3.
4.
5.
'''
# 一个简单示例2
def my_add(x,y):
    '''add function'''
    s = x+y
    return s

#这样子可以显示你的注释
print(my_add.__doc__)   
print(help(my_add))
```

```python
# 5.2函数参数
'''
最常用的是位置参数是最常用的一种参数形式
    调用函数时实参和形参的数量必许相同，且位置必须是一致的
'''
def my_add2(x,y):
    '''add function'''
    print("x="+str(x),"y="+str(y))
    s = x+y
    return s

print(my_add2("hello","world"))

```

```python
'''
第二种是默认值参数
def 函数名(...,形参名=默认值):
    函数体

注意 默认值参数右边不能再次出现形参 一定要放在最后
后续也可以设置实参 最后函数还是会调用实参来运算
'''
def my_add3(x,y=10):
    '''add function'''
    s = x+y
    return s
```

```python
'''
关键词参数
这个是在函数调用的时候使用的 可以通过 形参名=值 的形式传递参数 称为关键词参数
'''
def my_add4(x,y):
    '''add function'''
    s = x+y
    return s

print(my_add4(y=6,x=9))
# 不过你这样写 就一定要注意规范
```

```python
# 例5-1
# 定义函数 计算一组成绩的平均值 调用函数
def cal_avg(alist):
    s=sum(alist)
    n=len(alist)
    avg=round(s/n,1)
    return avg

scores=[30,65,69,42,96]
result=(cal_avg(scores))
print(f"平均分为:{result:.2f}")

```

```python
# 例子5-3 定义函数 可以按照指定字符和行数打印
def print_fig(char,rows=10):
    '''打印图形'''
    for i in range(rows+1):
        print(char * i)

print_fig("*")
print_fig('哈')
```

```python
'''
可变长度的参数列表
调用函数时 可以传递不同数目的实参
*arg 表示元组的变长参数 可以让不定长度的参数变成元组传递给函数
**arg 表示字典变长参数 传入不定长度的键值对
'''

def my_sum(*nums):
    print('nums:',nums)
    s=0
    for n in nums:
        s+=n
        return s
    
def my_sum2(**nums):
    print('nums:',nums)
    s=0
    for key in nums:
        s+=nums
        return s
    
my_sum(4,6)
my_sum2(a=4,b=5,c=5,d=5,e=5,f=9)
    
```

```python
# 5.3函变量的作用域
# 指的是变量起作用的代码范围
# 函数内部起作用：局部变量
# 在函数外部定义的变量：全局变量
```

```python
def my_add5():
    '''add function'''
    local_num=3 #局部变量
    
    return global_num+local_num
global_num=5
print(my_add5())
print(global_num)
# print(local_num) 会报错
```

```python
# 当局部变量和全局变量重名了的时候 局部变量会在自己的作用域内部发挥作用 出来之后才会使用全局变量
def my_add6():
    x = 3
    return x+x

x=6
print(my_add6())
```

```python
#可以通过global关键词 可以在函数内定义或者使用全局变量
def my_add7():
    global x
    print(x)        #此处结果为5
    x=3
    return x+x

x=5

print(my_add7())    #结果为6
print(x)            #结果为3
```

```python
# 5.4lamdba表达式
# 格式(方括号可选可不选)
''' [函数名=] lambda 参数1,参数2,...,参数n:表达式 '''
# lambda表达式拥有自己的命名空间 不能访问自由参数列表之外或全局命名空间里的参数(外面的参数它接收不到)
# 相当于只有一条return语句的小函数 表达式的值作为函数的返回值
```

```python
my_add =lambda x,y:x+y
my_add(3,5)
# my_add('hello','world')
```

```python
# lambda尤其适合需要一个函数作为另一个函数参数的场合
#例5-5 lambda将一组字符串表示的商品编号 按照编号值的大小进行排列
product_ids = ['181','22','356','55']
sorted(product_ids)                         #由于你是字符串 我们默认是按张ASCALL码去排序
sorted(product_ids,key = lambda x:int(x))   #lambda表达式 先把数据转化为int数字 再进行数字大小的排序  但是他返回的结果还是字符串

# 这个很好用
```
